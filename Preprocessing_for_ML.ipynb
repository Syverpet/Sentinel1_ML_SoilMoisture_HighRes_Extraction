```python
empty_df1 = pd.DataFrame(columns =['Beta_VH_Div_VV', 'Beta_VH_Minus_VV','Beta_VH_multi_VV',
 'Beta_VV_Div_VH', 'Beta_VV_Minus_HV', 'Beta_VV_Plus_HV', 'Gamma_VH_Div_VV', 'Gamma_VH_Minus_VV',
 'Gamma_VH_multi_VV', 'Gamma_VV_Div_VH', 'Gamma_VV_Minus_HV', 'Gamma_VV_Plus_HV', 'Sigma_VH_Div_VV',
 'Sigma_VH_Minus_VV', 'Sigma_VH_multi_VV', 'Sigma_VV_Div_VH', 'Sigma_VV_Minus_HV', 'Sigma_VV_Plus_HV',
 'TFGamma_VH_Div_VV', 'TFGamma_VH_Minus_VV', 'TFGamma_VH_multi_VV', 'TFGamma_VV_Div_VH', 'TFGamma_VV_Minus_HV',
 'TFGamma_VV_Plus_HV', 'date', 'lon', 'lat', 'name', 'S1_track'])

dates_list = ['07Nov2021', '13Nov2021', '19Nov2021', '01Dec2021']
date_dict = {'07Nov2021':'07.11.2021', '13Nov2021':'13.11.2021', '19Nov2021':'19.11.2021', '01Dec2021':'01.12.2021'}

track_list = [52, 147]

for y in track_list:
    print(f'Track = {y}')
          
    for i in dates_list:


        All_fields_test1 = All_fields[(All_fields["date"] == i) & (All_fields["S1_track"] == y)]


        All_fields_test2 = All_fields_test1.pivot_table(index=['name'], 
                    columns=['expression_prod_type'], values='Value').rename_axis(None, axis='columns').rename_axis(None, axis='rows')

        All_fields_test2['date'] = i

        print(All_fields_test2.shape)

        df_lat_long = All_fields[(All_fields["date"] == '07Nov2021') & (All_fields["S1_track"] == y) & (All_fields["expression_prod_type"] == 'TFGamma_VV_Plus_HV')]


        All_fields_test2['lon'] = df_lat_long['lon'].tolist()
        All_fields_test2['lat'] = df_lat_long['lat'].tolist()
        All_fields_test2['name'] = df_lat_long['name'].tolist()
        All_fields_test2['S1_track'] = df_lat_long['S1_track'].tolist()
        
        
        df_SM_data2 = df_SM_data[(df_SM_data['Date'] == date_dict[i])]


        print(All_fields_test2.shape)
        print(df_SM_data2.shape)

        All_fields_test3 = pd.merge(All_fields_test2, df_SM_data2, on="name")


        empty_df1 = empty_df1.append(All_fields_test3)

        print(empty_df1.shape)
    


```

```python
# reverse dataframe
empty_df2 = empty_df1.iloc[:, ::-1]


# Split dataframes on tracks 
track52 = empty_df2[empty_df2['S1_track'] == 52]

track147 = empty_df2[empty_df2['S1_track'] == 147]


#Columns for splitting
SAR_cols = ['TFGamma_VV_Plus_HV', 'TFGamma_VV_Minus_HV', 'TFGamma_VV_Div_VH', 'TFGamma_VH_multi_VV',
         'TFGamma_VH_Minus_VV', 'TFGamma_VH_Div_VV', 'Sigma_VV_Plus_HV', 'Sigma_VV_Minus_HV', 'Sigma_VV_Div_VH',
         'Sigma_VH_multi_VV', 'Sigma_VH_Minus_VV', 'Sigma_VH_Div_VV',
         'Gamma_VV_Plus_HV', 'Gamma_VV_Minus_HV', 'Gamma_VV_Div_VH', 'Gamma_VH_multi_VV', 'Gamma_VH_Minus_VV',
         'Gamma_VH_Div_VV', 'Beta_VV_Plus_HV', 'Beta_VV_Minus_HV', 'Beta_VV_Div_VH', 'Beta_VH_multi_VV',
         'Beta_VH_Minus_VV', 'Beta_VH_Div_VV']

non_SAR_cols = ['target', 'std', 'Field', 'Date', 'slope', 'nr_dem', 'nr_', 'name', 'lon', 'lat_s2', 'lat', 'index',
         'elev', 'date', 'aspect', 'TRI', 'TPI',
                'Wstdev', 'Wmedian', 'Wmean', 'Vstdev', 'Vmedian', 'Vmean', 'Imedian', 'Imean', 'Istdev']


# Split columns of dataframes
track52_2 = track52[SAR_cols]
track147_2 = track147[SAR_cols]

non_SAR_records = track52[non_SAR_cols]

# add sifix:
track52_2.columns = [(x + '_52') for x in track52_2.columns]
track147_2.columns = [(x + '_147') for x in track147_2.columns]


# concat 52 and 147 tracks togheter
merged_tracks = pd.concat([non_SAR_records, track52_2, track147_2], axis=1)

merged_tracks
```

```python
merged_tracks['fold_group'] = merged_tracks.date.str[:-7]

merged_tracks['name'] = merged_tracks.index

All_data_final = merged_tracks.drop(columns = ['lat_s2', 'index', 'date', 'nr_', 'nr_dem'])





All_data_final.to_csv('C:/Users/Laptop/Documents/1. Thesis 2021/SAR/Extract_for_ML_trial1/ML_Analysis/Final_data_for_analysis.csv')

All_data_final

```
