```python
# Mean function applied to pixels extracted
def Average(lst):
    return sum(lst) / len(lst)

import traceback
from snappy import GeoPos
from snappy import PixelPos
import pandas as pd
```

```python
list_of_stacks5 = ["52_stack_nov_sigma", "52_Beta_stack_nov", "52_Gamma_stack_nov", 
                  "147_Beta_stack_nov", "147_Gamma_stack_nov", "147_stack_nov_sigma", "147_TF_Gamma_nov", "52_TF_Gamma_stack_nov"]


S1_track_dict = {"52_TF_Gamma_stack_novBandMaths_16_12":'52', "52_stack_nov_sigmaBandMaths_16_12":'52', "52_Beta_stack_novBandMaths_16_12":'52',
                 "52_Gamma_stack_novBandMaths_16_12":'52', "147_Beta_stack_novBandMaths_16_12":'147', "147_Gamma_stack_novBandMaths_16_12":'147', 
                   "147_TF_Gamma_novBandMaths_16_12":'147', "147_stack_nov_sigmaBandMaths_16_12":'147'}

dict_for_stacks2 = {"52_TF_Gamma_stack_novBandMaths_16_12":"TFGamma", "52_stack_nov_sigmaBandMaths_16_12":"Sigma", 
                    "52_Beta_stack_novBandMaths_16_12":"Beta", "52_Gamma_stack_novBandMaths_16_12":"Gamma", 
                  "147_Beta_stack_novBandMaths_16_12":"Beta", "147_Gamma_stack_novBandMaths_16_12":"Gamma", 
                    "147_TF_Gamma_novBandMaths_16_12":"TFGamma", "147_stack_nov_sigmaBandMaths_16_12":"Sigma"}

A_field = pd.read_csv('C:/Users/Laptop/Documents/1. Thesis 2021/SAR/Extract_for_ML_trial1/Pixel_extracts_results/A_Field.csv')
B_field = pd.read_csv('C:/Users/Laptop/Documents/1. Thesis 2021/SAR/Extract_for_ML_trial1/Pixel_extracts_results/B_Field.csv')

# DEFINE THE FIELD
field = 'A'
field_cords = A_field

Name_add = 0 # For A

#Name_add = 30 # For B


for w in list_of_stacks5:
    
    path_to_bandsmaths_data = f'C:/Users/Laptop/Documents/1. Thesis 2021/SAR/Extract_for_ML_trial1/BandMaths_results/{w}BandMaths_16_12.dim'

    product_subset = ProductIO.readProduct(path_to_bandsmaths_data)
    prod_name = product_subset.getName()
    print(f'STARTING: {prod_name} ...')
    
    
    bands_names = list(product_subset.getBandNames())[0:24]
    gc = product_subset.getSceneGeoCoding()

    empty_df = pd.read_csv('C:/Users/Laptop/Documents/1. Thesis 2021/SAR/Extract_for_ML_trial1/Pixel_extracts_results/empty_df.csv')
    empty_df2 = empty_df.iloc[1: , 1:]
    empty_df3 = empty_df.iloc[1: , 1:]

    coord_list_nr = list(range(30))


    for y in coord_list_nr:

        lat, lon = (field_cords.loc[y][1], field_cords.loc[y][2])
        name = int(field_cords.loc[y][0])+Name_add
        
        #print(bands_names)

        cols = ['lat', 'lon', 'X', 'Y']
        cols.extend(bands_names)
        
        df_bands = pd.DataFrame(columns=cols)
        
        

        pixel_pos = gc.getPixelPos(GeoPos(lat, lon), None)
        data = list()

        data.append(str(name))
        data.append(lat)
        data.append(lon)
        data.append(int(pixel_pos.x))
        data.append(int(pixel_pos.y))

        for i, band_name in enumerate(bands_names):
            temp_band = product_subset.getBand(band_name)
            width, height = temp_band.getRasterWidth(), temp_band.getRasterHeight()
            try:
                tmp = np.zeros(3*3)
                temp_band.readPixels(int(pixel_pos.x-1), int(pixel_pos.y-1), 3, 3, tmp) #Pixels extracted
                
                tmp2 = [x for x in tmp if str(x) != 'nan']
                average_tmp = Average(tmp2) #calc mean of pixel list
                data.append(average_tmp)


            except Exception as e:
                #print(band_name)
                #print(width, height)
                #print(int(pixel_pos.x), int(pixel_pos.y))
                #print(e)
                traceback.print_exc()

        
        empty_df2.loc[df_bands.shape[0]] = data
        empty_df3 = empty_df3.append(empty_df2, ignore_index=True)

    empty_df4 = empty_df3.drop(columns=['X', 'Y'])
    empty_df5 = empty_df4.melt(id_vars=['name', "lat", "lon"], var_name="BandsDates", value_name="Value")
    empty_df5['V_band'] = empty_df5.BandsDates.str[:-9]
    empty_df5['date'] = empty_df5.BandsDates.str[-9:]
    empty_df5['Field'] = field
    empty_df5['S1_track'] = S1_track_dict[prod_name]
    empty_df5['prod_type'] = dict_for_stacks2[prod_name]
    empty_df5['expression'] = empty_df5.V_band.str[7:]
    
    print(empty_df5.shape)
    print(f'DONE: {prod_name}!')
    
    empty_df5.to_csv(f'C:/Users/Laptop/Documents/1. Thesis 2021/SAR/Extract_for_ML_trial1/Pixel_extracts_results/PE_{w}_17_12_{field}_field.csv')

```

```python
def plotBand(product, band, vmin, vmax):
    band = product.getBand(band)
    w = band.getRasterWidth()
    h = band.getRasterHeight()
    print(w, h)
    band_data = np.zeros(w * h, np.float32)
    band.readPixels(0, 0, w, h, band_data)
    band_data.shape = h, w
    width = 12
    height = 12
    plt.figure(figsize=(width, height))
    imgplot = plt.imshow(band_data, cmap=plt.cm.binary, vmin=vmin, vmax=vmax)
    return imgplot

#width = product.getSceneRasterWidth()
#print("Width: {} px".format(width))
#height = product.getSceneRasterHeight()
#print("Height: {} px".format(height))
#band_names = product.getBandNames()
#print("Band names: {}".format(", ".join(band_names)))
#band = product.getBand(band_names[0])
#print(band.getRasterSize())
plotBand(product, "Gamma0_VH_mst_07Nov2021", 0, 1)
```
